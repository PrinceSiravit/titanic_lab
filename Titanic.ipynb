### 3. Delete all rows where `Embarked` is empty.

```python
#Use dropna method for delete all rows in the Empty column which is empty.
df.dropna(subset = ['Embarked'])
```

### 4. Fill all empty cabins with ¯\(ツ)/¯

## Step 3: Feature extraction

### 1. Create `FamilyCount` from sum of `SibSp` and `Parch`

```python
# Create FamilyCount column from sum of two columns 
df['FamilyCount'] = df['SibSp'] + df['Parch']
```

```python
df
```

### 2. Create a column `IsReverend` 

```python
#Create a new boolean column from value in Name columns. 
#If the value in column 'Name' has rev in their name, it will replace True in the new columns. 
#If not, replace a Flase
df['IsReverend'] = df['Name'].str.contains('Rev')
```

```python
df
```

```python
#Change type of value in IsReverend column to be integer.
df['IsReverend'] = df['IsReverend'].astype(int)
```

```python
df
```

### 3. Convert categories

```python
#Use get_dummies to create new columns follow each value in that column.
dum = pd.get_dummies(df['Embarked'])
dum
```

```python
#Merge two dataframe by concat method.
df = pd.concat((df, dum), axis = 1)
```

```python
df
```

```python
#Rename columns by rename method
df.rename(columns = {'C': 'Embarked_C', 
                     'Q': 'Embarked_Q', 'S': 'Embarked_S' },inplace = True)
```

```python
#Create dummy variable columns from 'sex' column.
sex = pd.get_dummies(df['Sex'])
```

```python
sex.head()
```

```python
#Merge two dataframe
df = pd.concat((df,sex), axis =1)
```

```python
df
```

### Extract Title

```python
#Use split method for split string value after a full stop and remove it by strip method. 
df['title'] = df['Name'].apply(lambda x : x.split('.')[0].strip())
```

```python
df
```

```python
#Use split method for split string value before a comma and remove it by strip method.
df['title_new'] = df['title'].apply(lambda x : x.split(',')[1].strip())
```

```python
df
```

```python
#Drop column
df.drop(['title'], axis = 1, inplace = True)
```

```python
df
```

```python
#Create dummy variable columns by get_dummies function
title = pd.get_dummies(df['title_new'])
```

```python
#Merge two dataframes by concat method
df = pd.concat((df,title), axis =1)
```

```python
df.head()
```

## Exploratory Data Analysis

#### What was the survival rate overall?

```python
#Finding % of people who survived in Survived column by using mean function
df['Survived'].mean()
```

Ans: The survival rate overall is around 38% 

#### Which gender fared the worst? What was their survival rate?

```python
#Finding an average of each type which grouped by Sex column
df.groupby('Sex')['Survived'].mean()
```

Ans: The male gender has the worst one which is around 19% 

#### What was the survival rate for each Pclass?

```python
#Finding a survival of each group which grouped by Pclass
df.groupby('Pclass')['Survived'].mean()
```

Ans: The upper class is around 63%, the middle class is around 47%, and the lower class is around 24%

#### Did any reverends survive? How many?

```python
#Finding the number of people who are reverend.
df.groupby('IsReverend')['Survived'].sum()
```

Ans: No reverend survived. 

#### What is the survival rate for cabins marked ¯\(ツ)/¯

```python
#Finding the survival percentage of people who are cabin which has marked.
df.groupby(df['Cabin'] ==  '¯\(ツ)/¯')['Survived'].mean()
```

Ans: The survival rate of cabin who mask ¯\(ツ)/¯ is around 30%

#### What is the survival rate for people whose Age is empty?

```python
#Finding an average of empty value in Age column.
df.groupby(df['Age'] == '')['Survived'].mean()
```

Ans: The survival rate of people has not information about their ages is around 62% (1-0.38)

#### What is the survival rate for each port of embarkation?

Ans: The survival of each port of embarkation are around 55% of Cherbourg port, around 39 of Queenstown port, and around 34% of Southampton port.

## Plotting 

```python
#Plotting a survival rate graph by port and sex types.
sns.barplot(data=df, x = 'Survived', y = 'Embarked', hue='Sex', ci= None);
```

Ans: From the graph, it shows the survival rate of each sex which will embark on each port. For people who will embark on Cherbourg port, they will have the highest a survival rate follow by Southampton and Queenstown ports which female has higher survival rate than male in every ports.

```python
#Plotting a survival rate graph by each rank of socio-economic status and sex types. 
sns.barplot(data=df, x = 'Pclass', y = 'Survived', hue = 'Sex', palette = 'magma', ci = None);
```

Ans: From the graph, it shows the survival rate of each proxy for socio-economic status(SES) which consist of 1 = Upper, 2 = Middle, and 3 = Lower. For people who are Upper SES, they will have the highest a survival rate follow by Middle and Lower SES which female has higher survival rate than male in every status. It can be the evidence of inequality in social obviously.
